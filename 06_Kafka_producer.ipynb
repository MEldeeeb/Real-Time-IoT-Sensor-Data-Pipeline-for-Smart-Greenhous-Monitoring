```python
def create_kafka_producer():
    try:
        producer = KafkaProducer(bootstrap_servers = 'localhost:9092')
        return producer
    except:
        print("NoBrokersAvailable")

def publish_message(topic_name,message, kafka_producer):
    message = str(message)
    encoded_message = bytearray(message.encode("utf-8"))
    kafka_producer.send(topic_name,encoded_message)
```

```python
# create kafka produces 
producer = create_kafka_producer()
```
